```python
import pandas as pd
import requests 
import json

#Visualizations on map
import folium
from folium.plugins import MarkerCluster
```

```python
df = pd.read_csv("Restaurants.csv")
```

```python
df.head()
```

```python
for i,row in df.iterrows():
    Address = str(df.at[i,'street'])+','+ str(df.at[i,'zip'])+','+ str(df.at[i,'city'])+','+ str(df.at[i,'country'])
    parameters = {

        "key" : "29c1LGWSagAgRW7SHIb5pJtrjJrTQaFd",
        "location" : Address
    }

    response = requests.get("http://www.mapquestapi.com/geocoding/v1/address", params = parameters)
    data = json.loads(response.text)['results']

    latitude = data[0]['locations'][0]['latLng']['lat']
    longitude = data[0]['locations'][0]['latLng']['lng']
    
    df.at[i,'latitude']=latitude
    df.at[i,'longitude']=longitude
    
    
```

```python
df.head(5)
```

```python
maplocn = folium.Map(location=[52.52569,13.36846],tiles='OpenStreetMap',zoom_start=5)

for j,row in df.iterrows():
    lat = df.at[j,'latitude']
    long = df.at[j,'longitude']
    rest = df.at[j,'restaurant']
    
    if rest == 'McDonalds':
        color='blue'
    else: 
        color='red'
    popup = str(df.at[j,'restaurant'])+'<br>'+str(df.at[j,'zip'])+'<br>'+ str(df.at[j,'city'])+'<br>'+ str(df.at[j,'country'])
    folium.Marker(location=[lat,long],popup=popup,icon=folium.Icon(color=color)).add_to(maplocn)
```

```python
maplocn.save("locn.html")
```
